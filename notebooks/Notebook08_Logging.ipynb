```python
from datetime import datetime
from pyspark.sql import Row
```

```python
tables = [
    "biotech_funding_silver",
    "clinical_trials_silver",
    "disease_burden_silver",
    "drug_approvals_silver",
    "pharma_companies_silver"
]
```

```python
log_data = []
```

```python
for table in tables:

    start_time = datetime.now()

    try:

        df = spark.table(f"workspace.default.{table}")

        row_count = df.count()

        status = "SUCCESS"

        error_message = ""

    except Exception as e:

        row_count = 0

        status = "FAILED"

        error_message = str(e)

    end_time = datetime.now()

    log_data.append(
        Row(
            table_name=table,
            start_time=str(start_time),
            end_time=str(end_time),
            row_count=row_count,
            status=status,
            error_message=error_message
        )
    )
```

```python
log_df = spark.createDataFrame(log_data)
```

```python
display(log_df)
```

```python
log_df.write \
.format("delta") \
.mode("append") \
.saveAsTable("workspace.default.pipeline_logs")
```
