```python
file_path = "/Volumes/dbacademy/default/raw_data/biotech_funding.csv"

df = spark.read.csv(
    file_path,
    header=True,
    inferSchema=True
)
```

```python
df.count()
```

```python
len(df.columns)
```

```python
df.columns
```

```python
from pyspark.sql.functions import col, sum
```

```python
df.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in df.columns
]).display()
```

```python
df.count()
```

```python
df.distinct().count()
```

```python
from pyspark.sql.functions import count
```

```python
df.groupBy("deal_id").count().display()
```

```python
df.groupBy("deal_id") \
  .count() \
  .filter("count > 1") \
  .display()
```
