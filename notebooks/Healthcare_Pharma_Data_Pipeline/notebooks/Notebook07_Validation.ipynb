```python
tables = [
    "biotech_funding_silver",
    "clinical_trials_silver",
    "disease_burden_silver",
    "drug_approvals_silver",
    "pharma_companies_silver"
]
```

```python
for table in tables:

    print("=" * 80)

    count = spark.table(
        f"workspace.default.{table}"
    ).count()

    print(f"{table}")

    print(f"Row Count : {count}")
```

```python
for table in tables:

    print("=" * 80)

    print(table)

    spark.table(
        f"workspace.default.{table}"
    ).printSchema()
```

```python
for table in tables:

    df = spark.table(f"workspace.default.{table}")

    total = df.count()

    distinct = df.distinct().count()

    print("=" * 80)

    print(table)

    print("Total Rows :", total)

    print("Distinct Rows :", distinct)

    print("Duplicate Rows :", total - distinct)
```

```python
from pyspark.sql.functions import *

for table in tables:

    print("=" * 80)

    print(table)

    df = spark.table(f"workspace.default.{table}")

    display(
        df.select(
            [
                sum(col(c).isNull().cast("int")).alias(c)
                for c in df.columns
            ]
        )
    )
```
